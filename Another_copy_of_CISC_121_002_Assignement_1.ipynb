I reviewed the top solution under leetcode solutions tab https://leetcode.com/problems/longest-common-prefix/solutions/3273176/python3-c-java-19-ms-beats-99-91/ where they do this by sorting approach ; then I watched another youtube video https://www.youtube.com/watch?v=0sWShKIJoo4 where they use my approach . for me it was easy to understand and debugg.

```python
class Solution:
    def longestCommonPrefix(self, strs: list[str]) -> str:
        if not strs:
            return "" #Handle edge case: if the list is empty, return empty string
        res = ""  #result string to store common prefix
        min_len = min(len(s) for s in strs) #find the length of the shortest string to avoid index out of bounds
        for i in range(min_len): #iterate through each character position
            for j in range(len(strs)):
                if strs[j][i] != strs[0][i]:
                    return res
            res += strs[0][i]
        return res






```

```python
class Solution:
    def longestCommonPrefix(self, strs: list[str]) -> str:
        res = "" #result string to store common prefix
        for i in range(len(strs[0])) : #iterate through each character position in the first string
            for s in strs: #check this character position across all strings
                if i == len(s) or s[i] != strs[0][i]: #two termination condition : if we reached the end of the string 's' (i==len(s)) or character mismatch found  (s[i] != strs[0][i])
                    return res #return the prefix
            res += strs[0][i] #add matching character
        return res #return the complete common prefix

```

```python
class Solution:
    def addDigits(self, num: int) -> int:
        while num > 10:  #continue until we have a single digit #misses the case if num==10
            digit_sum = 0 #reset sum for current iteration
            while num > 0: #extract and sum all digits of current number
                digit_sum += num % 10 #add rightmost digit to the sum
                num /= 10 #remove rightmost digit #true division float , loop never reaches zero
            num = digit_sum #update num with digit_sum for next iteration
        return num #return final single digit






```
